<h3>1.PREPROCESSING DATA</h3>

```python
# import libraries
import numpy as np
import math
import re
import pandas as pd
```

```python
# open file to take data
with open("/Users/Admin/DSLab Training Session/Session 1/data/deathRate.txt") as f:
    # read data, split and store in list
    data = f.read().splitlines()
    matrix=[]
    for line in data:
        # replace multiple white spaces by a white space
        row = re.sub(' +',' ',line)
        # delete the first element in each row
        row = row.split()[1:]
        matrix.append(row)
```

```python
# assign columns name
columns = np.append(["A" + str(i) for i in range(1,16)],"B")
# create DF
data = pd.DataFrame(matrix,columns=columns)
# index of data
data.index = [i for i in range(1, len(matrix)+1)]
# change datatype of data from string to number
data.apply(pd.to_numeric)
# some informations of data set
data.describe()
```
