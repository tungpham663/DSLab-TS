<h3>2. Ridge Regression</h3>

```python
# normalize the data
def normalize_data_and_add_ones(X):
    X=np.array(X)
    #search in each row to find min/max element in column
    X_max = np.array([[np.amax(X[:,column_id]) for column_id in range(X.shape[1])] for _ in range(X.shape[0])])
    X_min = np.array([[np.amin(X[:,column_id]) for column_id in range(X.shape[1])] for _ in range(X.shape[0])])
    #normalize Min-Max features scaling
    X_normalized = (X-X_min)/(X_max-X_min)
    ones=[[1] for _ in range(X.shape[0])]
    #return the X_normalize with the first col is 1
    return np.column_stack((ones,X_normalized))
```

```python
#normalize data
data=normalize_data_and_add_ones(data)
data
```

```python
# build class RidgeRegression
class RidgeRegression:
    def __init__(self):
        return
    
    # determine the optimal w
    def fit(self,X_train,Y_train,LAMBDA):
        assert len(X_train.shape)==2 and \
               X_train.shape[0]==Y_train.shape[0]
        # use the formula of optimal w: w* = tranpose(X^T.X).dot(X^T.Y)
        W = np.linalg.inv(
            X_train.transpose().dot(X_train) + 
            LAMBDA*np.identity(X_train.shape[1])
            ).dot(X_train.transpose()).dot(Y_train)
        return W
    
    # find the optimal w by gradient descent
    def fit_gradient_descent(self,X_train,Y_train,LAMBDA,learning_rate=0.01,max_num_epoch=100,batch_size=10):
        #initialize W with sample standard normal distribution
        W = np.random.randn(X_train.shape[1]).reshape(-1,1)
        last_loss=10e+8
        for ep in range(max_num_epoch):
            #array of index of X_train
            arr = np.array(range(X_train.shape[0]))
            #shuffle the order of elements of array
            np.random.shuffle(arr)
            #get the shuffle X_train and relative Y_train
            X_train = X_train[arr]
            Y_train = Y_train[arr]
            total_minibatch = int(np.ceil(X_train.shape[0]/batch_size))
            for i in range(total_minibatch):
                index = i*batch_size
                X_train_sub = X_train[index:index + batch_size]
                Y_train_sub = Y_train[index:index + batch_size]
                #compute gradient
                gradient = X_train_sub.T.dot(X_train_sub.dot(W) - Y_train_sub) + LAMBDA * W
                #update the value of w
                W = W - learning_rate * gradient
            new_loss = self.compute_RSS(self.predict(W,X_train),Y_train)
            #check the difference between current loss and previous loss is small enough
            if np.abs(new_loss-last_loss)<=1e-5:
                break
            last_loss=new_loss
        return W
    
    # predict the value of Y_new from X_new and w
    def predict(self,W,X_new):
        X_new = np.array(X_new)
        Y_new = X_new.dot(W)
        return Y_new
    
    # compute RSS of y_predicted and y_new
    def compute_RSS(self,Y_new,Y_predicted):
        # the average of square of Y_new-Y_predicted
        loss = 1. / Y_new.shape[0] * \
               np.sum((Y_new-Y_predicted)**2)
        return loss
    
    def get_the_best_LAMBDA(self,X_train,Y_train):
        # compute the average RSS given LAMBDA, num_folds by cross-validation
        def cross_validation(num_folds,LAMBDA):
            row_ids = np.array(range(X_train.shape[0]))
            # np.split() requires equal divisions
            valid_ids = np.split(row_ids[:len(row_ids) - len(row_ids)%num_folds],num_folds)
            # append the remainder of row_ids into the last element of valid_ids
            valid_ids[-1] = np.append(valid_ids[-1],row_ids[len(row_ids) - len(row_ids)%num_folds:])
            # the list of D/Di for i in range(num_folds)
            train_ids = [[k for k in row_ids if k not in valid_ids[i]] for i in range(num_folds)]
            aver_RSS=0
            for i in range(num_folds):
                valid_parts = {'X':X_train[valid_ids[i]],'Y':Y_train[valid_ids[i]]}
                train_parts = {'X':X_train[train_ids[i]],'Y':Y_train[train_ids[i]]}
                W = self.fit(train_parts['X'],train_parts['Y'],LAMBDA)
                Y_predicted = self.predict(W,valid_parts['X'])
                aver_RSS += self.compute_RSS(valid_parts['Y'],Y_predicted)
            return aver_RSS/num_folds
        
        def range_scan(best_LAMBDA,minimum_RSS,LAMBDA_values):
            #loop in lambda_values set
            for current_LAMBDA in LAMBDA_values:
                #caluculate average RSS with each lambda
                aver_RSS = cross_validation(num_folds=5,LAMBDA = current_LAMBDA)
                #get the best value of lambda
                if aver_RSS < minimum_RSS:
                    best_LAMBDA = current_LAMBDA
                    minimum_RSS = aver_RSS
            return best_LAMBDA, minimum_RSS
        
        best_LAMBDA, minimum_RSS = range_scan(best_LAMBDA=0,minimum_RSS=10e+8,
                                              LAMBDA_values = range(60)) #[0,1,2,...,59]
        
        LAMBDA_values = [k * 1/1000 for k in range(
            max(0,(best_LAMBDA-1)*1000),(best_LAMBDA+1)*100,1)] # step size is 0.001
        
        best_LAMBDA, minimum_RSS = range_scan(best_LAMBDA=best_LAMBDA,minimum_RSS=minimum_RSS,LAMBDA_values=LAMBDA_values)
        return best_LAMBDA
```

```python
# split training set and test set(50/10)
X_train, Y_train = data[:50, :data.shape[1]-1], data[:50,data.shape[1]-1].reshape(-1,1)
X_test, Y_test = data[50:,:data.shape[1]-1], data[50:,data.shape[1]-1].reshape(-1,1)
#apply ridge regression
RR = RidgeRegression()
#get the optimal lambda
best_lambda=RR.get_the_best_LAMBDA(X_train,Y_train)
#W by fit by gradient descent algorithm
W_gd = RR.fit(X_train,Y_train,best_lambda)
Y_predicted = RR.predict(W_gd,X_test)
#return RSS
RR.compute_RSS(Y_test,Y_predicted)
```
